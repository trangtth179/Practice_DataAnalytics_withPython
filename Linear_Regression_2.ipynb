```python
#Import libraby
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#Load data
income = np.array([8,9,10,11,12,15,15,16,17,18,18,20,24,25,25]).reshape((-1,1))
expenditure = np.array([6,7,9,9,10,12,11,13,13,15,14,16,18,20,19])
```

```python
#Creat model
model = LinearRegression().fit(income, expenditure)
```

```python
#Get result
R_square = model.score(income, expenditure)
print(R_square)
intercept = model.intercept_
print(intercept)
slope = model.coef_
print(slope)
```

```python
plt.scatter(income, expenditure)
plt.plot(income, intercept + slope*income)
```

```python
pre_values = intercept + slope*income
plt.plot(income, expenditure, color = 'r', label = 'Actual', marker = 'o')
plt.plot(income, pre_values, color = 'b', label = 'Predict', marker = 'x')
plt.legend()
plt.grid()
```

```python
income_2 = np.array([23,26,28]).reshape((-1,1))
pre_values_2 = intercept + slope*income_2
print(pre_values_2)
```
