```python
#Import library
from scipy import stats
import matplotlib.pyplot as plt
```

```python
#Load data
income = [8,9,10,11,12,15,15,16,17,18,18,20,24,25,25]
expenditure = [6,7,9,9,10,12,11,13,13,15,14,16,18,20,19]
```

```python
#Create model
slope, intercept, r, p, std_err=stats.linregress(income, expenditure)
def func(x):
    return intercept + slope*x
model = list(map(func, income))
```

```python
#Visualization
plt.scatter(income,expenditure)
plt.xlabel("Income")
plt.ylabel("Expenditure")
plt.plot(income, model)
plt.show()
```

```python
#Pridict
pre_values = []
for i in income:
    pre_values.append(func(i))
    
plt.plot(income, expenditure, color = 'r', label = 'Actual', marker = 'o')
plt.plot(income, pre_values, color = 'darkgreen', label = 'Pridict', linestyle= '--', marker ='x')
plt.legend()
```
